```python
from faker import Faker
import json
import random
import requests
import datetime

# List of available languages
languages = ['vi_VN', 'zh_CN', 'ja_JP', 'ko_KR', 'th_TH', 'ne_NP', 'bn_BD', 'hi_IN']

# load flights data
with open('flights.json', 'r') as f:
    flights_data = json.load(f)

# load flights data
with open('passengers.json', 'r') as f:
    passengers_data = json.load(f)
    
# load flights data
with open('bookings.json', 'r') as f:
    bookings_data = json.load(f)
    
# load flights data
with open('payment_methods.json', 'r') as f:
    payment_methods_data = json.load(f)
    
#
with open('seats.json', 'r') as f:
    seats_data = json.load(f)

# Generate passenger data for each language
international_bookings = []
for language in languages:
    fake = Faker(language)
    for i in range(1000):
        # generate a random booking code
        booking_code = random.choice(bookings_data)
        # select a random flight
        flight = random.choice(flights_data)
        # generate a random passenger ID
        passenger_id = random.choice(passengers_data)
        # generate a random total amount
        total_amount = random.choice(bookings_data)
        payment_methods = random.choice(payment_methods_data)
        seat_number = random.choice(seats_data)
        passport = random.choice(passengers_data)
        
        international_booking_data = {
            "bookingCode": booking_code['bookingCode'],
            "flightCode": flight['flightCode'],
            "passengerId": passenger_id['passengerId'],
            "totalAmount": total_amount['totalAmount'],
            "paymentMethod": payment_methods['paymentMethod'],
            "seatNumber": seat_number['seatNumber'],
            "passportNumber": passport['passportNumber'],
            "visaNumber": fake.random_int(min=100000000, max=999999999),
        }
        international_bookings.append(international_booking_data)

with open('international_bookings.json', 'w') as f:
    json.dump(international_bookings, f, indent=4)
```
